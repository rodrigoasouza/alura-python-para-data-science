## Parameters, Conditionals, and Type Conversion

```python
def apply_for_drivers_license():
  old = input('How old are you? ')
  old = int(old)

  if old >= 15 :
    print('Can apply')
  else :
    print('Can\'t apply')
```

```python
apply_for_drivers_license()
```
