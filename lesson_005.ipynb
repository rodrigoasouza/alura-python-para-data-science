## Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1, 7))
y = [9, 6, 7, 4, 7, 9]
```

```python
plt.title('Math notes')
plt.xlabel('Exams')
plt.ylabel('Grades')
plt.plot(x, y, marker='o')

plt.show()
```

### Logic and final value

```python
grades_math = ['Math', 8, 7, 6, 6, 7, 7, 8, 10]
grades_english = ['English', 9, 9, 9, 8, 5, 6, 8, 5]
grades_geography = ['Geography', 10, 10, 6, 7, 7, 7, 8, 7]

grades_list = [grades_math, grades_english, grades_geography]
```

```python
for grades in grades_list:
  x = list(range(1, 9))
  y = grades[1:]

  plt.plot(x, y, marker='o')
  plt.xlabel('Exams')
  plt.ylabel('Grades')
  plt.title(grades[0])
  plt.show()
```
